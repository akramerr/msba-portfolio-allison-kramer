## **1. Introduction**

### **1.1 Problem Statement**

New York City’s taxi system has long been one of the city’s most prominent modes of transportation. Greens taxis were launched with the purpose to serve the areas that are rarely covered by Yellow taxis. However, the factors driving changes in Green Taxi ridership and revenue patterns from 2019 to 2023 remain unexplored, particularly across neighborhoods such as Manhattan, Brooklyn, Queens, the Bronx, and Staten Island. These trends have also been impacted by major external factors, including the COVID-19 pandemic and the phased reduction of Green Taxi operations. Key questions regarding how passenger counts, trip distances, payment types, fare amounts, and tip amounts change over time and across geographic areas. Understanding these patterns is essential for taxi drivers, city officials, and urban planners who rely on accurate insights to guide traffic management, resource allocation, and transportation policy decisions.





### **1.2 Motivation**
Currently, NYC green taxis are being phased out, impacting revenue and urban mobility. NYC’s green taxi system illustrates how local demand patterns, economic disruptions such as COVID-19, and policy decisions directly shape urban mobility. Analyzing these trips allows us to transform transportation data into meaningful insights for stakeholders—including city planners, government officials, and taxi drivers. We care because we want to optimize current green taxi operations, and provide insights as green taxis potentially merge into other taxi services and rideshare platforms.

### **2.2 Handling spilled over data and date columns**
Upon data inspection, we learned that there are some old data spilled over to the year after. For instance, 2019 dataset should contain only 2019 data but has data from other year. To prevent duplication issue, we developed SQL query to remove any rows that do not belong in their respective year.

Additionally, changed `lpep_pickup_datetime` and `lpep_dropoff_datetime` date columns from string type to timestamp.

Similarly, we displayed code cell and result of 2019 dataset and hided the rest of the years.

```python
################# 2019 #####################
# VERIFY THAT THERE IS DATA FROM OTHER YEAR IN 2019 DATASET
%%bigquery
SELECT
  EXTRACT(YEAR FROM SAFE.PARSE_TIMESTAMP('%m/%d/%Y %I:%M:%S %p', lpep_pickup_datetime)) AS year,
  COUNT(*) AS cnt
FROM `ba-775-cs.775_datasets.2019_trip_records`
GROUP BY year
ORDER BY year;
```

```python
# FILTERED OUT NON-2019 DATA AND CONVERTED DATE STRING TO TIMESTAMP
%%bigquery
CREATE OR REPLACE TABLE `ba-775-cs.775_datasets.2019_trip_records_date_cleaned` AS
SELECT
  SAFE.PARSE_TIMESTAMP('%m/%d/%Y %I:%M:%S %p', lpep_pickup_datetime) AS pickup_ts,
  SAFE.PARSE_TIMESTAMP('%m/%d/%Y %I:%M:%S %p', lpep_dropoff_datetime) AS dropoff_ts,
  * EXCEPT(lpep_pickup_datetime, lpep_dropoff_datetime)
FROM `ba-775-cs.775_datasets.2019_trip_records`
WHERE EXTRACT(YEAR FROM SAFE.PARSE_TIMESTAMP('%m/%d/%Y %I:%M:%S %p', lpep_pickup_datetime)) = 2019; --THIS LINE USED TO REMOVE NON-2019 DATA
```

### **2.3 Find and Replace Missing Values**

After finishing filtering out each dataset with non-belonging data, we ran query below to display count of total missing values in each colum in a dataset. We applied this query to the rest of the datasets.

Moreover, we learned that each dataset has missing values in some of the columns except `pickup and dropoff locations` dataset. We handled missing values in each column of each dataset as below:

Column | Issue | Action
-------|-------|-------
**ehail_fee**|empty|drop
**store_and_fwd_flag**|not relevant|drop
**RatecodeID**|missing value|replace missing values 99
**passenger_count**|missing value|replace with mode value
**payment_type**|missing value|replace with 5

```python
################## 2019 ###################
%%bigquery
-- COUNT NULL VALUES IN EACH COLUMN
SELECT
  COUNTIF(vendorid IS NULL) AS vendorid_nulls,
  COUNTIF(pickup_ts IS NULL) AS pickup_ts_nulls,
  COUNTIF(dropoff_ts IS NULL) AS dropoff_ts_nulls,
  COUNTIF(store_and_fwd_flag IS NULL) AS store_and_fwd_flag_nulls,
  COUNTIF(RatecodeID IS NULL) AS RatecodeID_nulls,
  COUNTIF(PULocationID IS NULL) AS PULocationID_nulls,
  COUNTIF(DOLocationID IS NULL) AS DOLocationID_nulls,
  COUNTIF(passenger_count IS NULL) AS passenger_count_nulls,
  COUNTIF(trip_distance IS NULL) AS trip_distance_nulls,
  COUNTIF(fare_amount IS NULL) AS fare_amount_nulls,
  COUNTIF(extra IS NULL) AS extra_nulls,
  COUNTIF(mta_tax IS NULL) AS mta_tax_nulls,
  COUNTIF(tip_amount IS NULL) AS tip_amount_nulls,
  COUNTIF(tolls_amount IS NULL) AS tolls_amount_nulls,
  COUNTIF(ehail_fee IS NULL) AS ehail_fee_nulls,
  COUNTIF(improvement_surcharge IS NULL) AS improvement_surcharge_nulls,
  COUNTIF(total_amount IS NULL) AS total_amount_nulls,
  COUNTIF(payment_type IS NULL) AS payment_type_nulls,
  COUNTIF(trip_type IS NULL) AS trip_type_nulls,
  COUNTIF(congestion_surcharge IS NULL) AS congestion_surcharge_nulls
FROM `ba-775-cs.775_datasets.2019_trip_records_date_cleaned`;
```

```python
# HANDLING MISSING VALUES
%%bigquery
CREATE OR REPLACE VIEW
  `ba-775-cs.775_datasets.2019_trip_records_view` AS
WITH
  cte AS (
  SELECT
    passenger_count AS passenger_mode,
    COUNT(*) AS mode
  FROM
    `ba-775-cs.775_datasets.2019_trip_records_date_cleaned`
  WHERE
    passenger_count IS NOT NULL
  GROUP BY
    passenger_count
  ORDER BY
    mode DESC
  LIMIT
    1 )
SELECT
  COALESCE(RatecodeID, 99) AS RatecodeID_imputed, --imputed missing values with 99 as it means 'unknown' in data dictionary
  COALESCE(payment_type, 5) AS payment_type_imputed, --imputed missing values with 5 as it means 'unknown' in data dictionary
  COALESCE(passenger_count, (
    SELECT passenger_mode
    FROM cte
  )) AS passenger_count_imputed, --imputed missing values with mode
  * EXCEPT (ehail_fee, store_and_fwd_flag, RatecodeID, payment_type, passenger_count)
FROM
  `ba-775-cs.775_datasets.2019_trip_records_date_cleaned`;
```

### **3.3 Date and Travel Time Insights**
The next query looks at the **length of the trip by the time of day** in order to figure out which is the busiest.

```python
%%bigquery
--Query to find the longest average trip time by time of Day
WITH trip_data AS (
  SELECT
    CASE
      WHEN EXTRACT(HOUR FROM pickup_ts) BETWEEN 0 AND 5 THEN 'Early Morning'
      WHEN EXTRACT(HOUR FROM pickup_ts) BETWEEN 6 AND 11 THEN 'Morning'
      WHEN EXTRACT(HOUR FROM pickup_ts) BETWEEN 12 AND 17 THEN 'Afternoon'
      ELSE 'Evening'
    END AS time_of_day,
    TIMESTAMP_DIFF(dropoff_ts, pickup_ts, MINUTE) AS trip_length_minutes
  FROM `ba-775-cs.775_datasets.all_trip_records_view`
  WHERE pickup_ts IS NOT NULL
    AND dropoff_ts IS NOT NULL
    AND dropoff_ts > pickup_ts
)
SELECT
  time_of_day,
  AVG(trip_length_minutes) AS avg_trip_length_minutes,
  COUNT(*) AS total_trips
FROM trip_data
GROUP BY time_of_day
ORDER BY
  CASE time_of_day
    WHEN 'Early Morning' THEN 1
    WHEN 'Morning' THEN 2
    WHEN 'Afternoon' THEN 3
    WHEN 'Evening' THEN 4
  END;
```

**Findings:**

For the above code, we have set ‘buckets’ wherein we have classified the time at which the trips have occurred. The split being:


*   Early Morning- Midnight to 5 am
*   Morning - 6am to 11am
* Afternoon 12pm to 5 pm and
* Evening being the rest of the

From the results of the query, the afternoon seems to be the busiest time of day with a value of 22.61 minutes, this could be a result of more people such as schools getting over, tourists exploring the city and people leaving from their offices to ironically beat the ‘rush hour’ back home. This is supported with the fact that the most number of trips occur in the afternoon, which is a value of 3,895,364. To give some perspective, the lowest is the early morning, having only 762,175 trips during the 5 years. For Early morning and morning the values are quite close to one another 21.07 and 21.67 respectively. The period when cabs take the shortest time to reach their destination is the Evening with trips taking around 18.59 minutes which is surprising since the total number of trips is second highest, yet ~3 minutes faster than Morning which has ~50,000 less trips.


The following query looks at the **average trip length in minutes** grouped by the pickup location ID.

```python
%%bigquery
-- Query to find the average trip length by the Pickup location
WITH trip_lengths AS (
  SELECT
    PULocationID,
    AVG(TIMESTAMP_DIFF(dropoff_ts, pickup_ts, MINUTE)) AS avg_trip_length_min,
    COUNT(*) AS total_trips
  FROM `ba-775-cs.775_datasets.all_trip_records_view`
  WHERE pickup_ts IS NOT NULL
    AND dropoff_ts IS NOT NULL
    AND dropoff_ts > pickup_ts
  GROUP BY PULocationID
),
zones AS (
  SELECT DISTINCT
    location_id,
    borough,
    zone_location
  FROM `ba-775-cs.775_datasets.pickups_dropoffs_monthly_view`
)
SELECT
  t.PULocationID,
  z.borough,
  z.zone_location,
  t.avg_trip_length_min,
  t.total_trips
FROM trip_lengths t
LEFT JOIN zones z
  ON t.PULocationID = z.location_id
WHERE z.borough NOT IN ('EWR', 'Unknown')
ORDER BY t.avg_trip_length_min DESC;
```

**Findings:**

It also joins the two tables together on the location in order to find out which specific boroughs and zones have the longest average trip time. From the results of the query, we can see that the borough that had the longest time was Staten Island which occupied the top 9 spots in terms of average trip length. The longest time on average is 61.14 minutes (slightly over an hour of average travel time) from Mariners Harbor in Staten Island.
This can be attributed to its location and how far it is from other parts of New York City. On the other end of the spectrum is Brooklyn; zone location being Green-Wood Cemetery, where the quickest average time is 15.36 minutes. 8 of the 10 lowest time taking trips on average were in Manhattan.

The following query shows **average cost of taxi by location**

```python
%%bigquery
WITH
  location_avg AS (
  SELECT
    PULocationID AS location_id,
    AVG(total_amount) AS avg_total_cost,
    COUNT(*) AS trip_count
  FROM
    `ba-775-cs.775_datasets.all_trip_records_view`
  WHERE
    pickup_ts IS NOT NULL
    AND dropoff_ts IS NOT NULL
    AND dropoff_ts > pickup_ts
  GROUP BY
    location_id ),
  zones AS (
  SELECT
    DISTINCT location_id,
    zone_location,
    borough
  FROM
    `ba-775-cs.775_datasets.pickups_dropoffs_monthly` )
SELECT
  l.location_id,
  z.borough,
  z.zone_location,
  l.avg_total_cost,
  l.trip_count
FROM
  location_avg l
LEFT JOIN
  zones z
ON
  l.location_id = z.location_id
WHERE
  z.zone_location IS NOT NULL
  AND z.borough IS NOT NULL
  AND z.borough != 'EWR'
ORDER BY
  l.avg_total_cost DESC
LIMIT
  10;
```

**Findings：**

The most expensive pickup location is Rossville/Woodrow, with an average cost of about $94. The next highest-cost locations are all in Staten Island, including Bloomfield/Emerson Hill, Arden Heights, and Mariners Harbor. These areas consistently show higher average trip costs compared with other NYC neighborhoods.

The following query shows **average cost of taxi by weekend VS weekday**, attempting to find out if weekend trips are more expensive than weekday trips.

```python
%%bigquery
-- Compute average taxi cost for Weekend vs Weekday (Green Cab only)

SELECT
  CASE
    WHEN EXTRACT(DAYOFWEEK FROM t.pickup_ts) IN (1, 7) THEN 'Weekend'   -- Sunday=1, Saturday=7
    ELSE 'Weekday'
  END AS day_type,                    -- Classify each trip as Weekend / Weekday

  AVG(total_amount) AS avg_total_cost,  -- Total trip cost
  COUNT(*) AS trip_count              -- Number of trips

FROM `ba-775-cs.775_datasets.all_trip_records_view` AS t

-- Join lookup table, filtered to Green Cab pickup zones
LEFT JOIN (
    SELECT *
    FROM `ba-775-cs.775_datasets.pickups_dropoffs_monthly`
    WHERE pickup_dropoff = 'Pickup'    -- Pickup locations only related to costs
) AS z
ON t.PULocationID = z.location_id

GROUP BY
  day_type

ORDER BY
  avg_total_cost DESC;
```

**Findings：**

Based on Green Cab data, the average trip cost is very similar between weekdays and weekends. There is no major price difference between weekday and weekend trips. This suggests that Green Cab pricing is mainly driven by trip distance and location, not by the day of the week.

```python
%%bigquery
WITH base AS (
  SELECT
    pickup_ts,
    CASE payment_type_imputed
      WHEN 1 THEN 'Credit Card'
      WHEN 2 THEN 'Cash'
      WHEN 3 THEN 'No Charge'
      ELSE 'Other'
    END AS payment_category,
 CASE
      WHEN EXTRACT(YEAR FROM pickup_ts) <= 2019
        THEN 'Pre-COVID'
      WHEN EXTRACT(YEAR FROM pickup_ts) IN (2020, 2021)
        THEN 'During COVID'
      WHEN EXTRACT(YEAR FROM pickup_ts) IN (2022, 2023)
        THEN 'Post-COVID'
    END AS covid_period
  FROM `ba-775-cs.775_datasets.all_trip_records_view`
  WHERE EXTRACT(YEAR FROM pickup_ts) BETWEEN 2019 AND 2023
    AND payment_type_imputed IN (1, 2, 3)
),
agg AS (
  SELECT
    EXTRACT(YEAR FROM pickup_ts) AS year,
    covid_period,
    payment_category,
    COUNT(*) AS trip_count
  FROM base
  GROUP BY year, covid_period, payment_category
),
totals AS (
  SELECT
    year,
    covid_period,
    SUM(trip_count) AS total_trips
  FROM agg
  GROUP BY year, covid_period
)
SELECT
  a.year,
  a.covid_period,
  a.payment_category,
  a.trip_count,
  SAFE_DIVIDE(a.trip_count, t.total_trips) AS pct_of_trips
FROM agg a
JOIN totals t
  USING (year, covid_period)
ORDER BY
  year,
  covid_period,
  payment_category;
```

**Finding:**

Credit card payments consistently made up the majority of trips across all COVID periods, and their share actually increased from Pre-COVID to Post-COVID, rising from about 57% to over 66%. In contrast, cash usage dropped sharply during COVID and never recovered to pre-pandemic levels, suggesting a lasting shift toward contactless, card-based payments.

The Following Query Shows **Congestion Surcharge Revenue** in Each Borough

```python
%%bigquery
WITH zone_lookup AS (
  SELECT DISTINCT
    location_id,
    borough
  FROM `ba-775-cs.775_datasets.pickups_dropoffs_monthly_view`
),
trips AS (
  SELECT
    pickup_ts,
    DOLocationID,
    congestion_surcharge
  FROM `ba-775-cs.775_datasets.all_trip_records_view`
  WHERE EXTRACT(YEAR FROM pickup_ts) BETWEEN 2019 AND 2023
)
SELECT
  z.borough,
  SUM(t.congestion_surcharge) AS total_congestion_surcharge,
  COUNT(*) AS trip_count,
  AVG(t.congestion_surcharge) AS avg_surcharge
FROM trips t
JOIN zone_lookup z
  ON t.DOLocationID = z.location_id
WHERE z.borough NOT IN ('EWR', 'Unknown')
GROUP BY z.borough
ORDER BY total_congestion_surcharge DESC;
```

**Finding:**

Manhattan accounts for virtually all congestion surcharge revenue, generating over 4 million dollars - far higher than any other borough - because the majority of taxi activity and congestion occurs in the city’s core. In contrast, boroughs like Queens, Brooklyn, the Bronx, and Staten Island contribute only a tiny fraction, reflecting lower trip density and fewer trips passing through designated surcharge zones.

The Following Query Shows Difference in **Payment Types for Tips** to Green Taxi Drivers

```python
%%bigquery
WITH base AS (
  SELECT
    payment_type_imputed,
    tip_amount,
    total_amount,
    CASE payment_type_imputed
      WHEN 1 THEN 'Credit Card'
      WHEN 2 THEN 'Cash'
      WHEN 3 THEN 'No Charge'
      WHEN 4 THEN 'Dispute'
      --WHEN 5 THEN 'Unknown'
      ELSE 'Other'
    END AS payment_category,
    (tip_amount/total_amount)*100 as tip_percentage
  FROM `ba-775-cs.775_datasets.all_trip_records_view`
  WHERE EXTRACT(YEAR FROM pickup_ts) BETWEEN 2019 AND 2023
    AND total_amount > 0
    AND payment_type_imputed != 5
)
SELECT
  payment_category,
  AVG(tip_percentage) AS avg_tip_percentage,
  COUNT(*) AS trip_count
FROM base
GROUP BY payment_category
ORDER BY avg_tip_percentage DESC;
```

**Finding:**

Credit card trips have by far the highest average tip percentage at nearly 12%, confirming that riders overwhelmingly tip when paying electronically. In contrast, cash, dispute, and no-charge trips show near-zero tip percentages, reflecting the fact that cash tips are possibly not recorded and that non-standard payment types rarely include gratuities.

```python
%%bigquery
WITH
  all_records AS (
  SELECT
    *,
    DATE_TRUNC(pickup_ts, MONTH) AS trip_month
  FROM
    `ba-775-cs.775_datasets.all_trip_records_view` ),
  monthly AS (
  SELECT
    location_id,
    borough,
    metric_month_ts AS monthly_trips,
    trip_count
  FROM
    `ba-775-cs.775_datasets.pickups_dropoffs_monthly_view` ),
  all_data AS (
  SELECT
    a.pickup_ts,
    a.dropoff_ts,
    a.trip_month,
    b.borough,
    a.tip_amount,
  FROM
    all_records a
  LEFT JOIN
    monthly b
  ON
    a.DOLocationID = b.location_id )
SELECT
  borough,
  --trip_month,
  ROUND(AVG(tip_amount),2) as avg_tip_amount
FROM
  all_data
WHERE
  borough NOT IN ('EWR',
    'Unknown')
  AND borough IS NOT NULL
GROUP BY
  1
ORDER BY
  1
```
